__1.1.- EN PRIMER LUGAR: PARTIR EL CONJUNTO EN TRAIN/TEST__ : 

Voy a separar el conjutno de datos en __TRAIN/TEST__, como me pide el ejercicio, pero a su vez: 
- En el mismo código voya a entrenar a un modelo para que me permita realizar predicciones. 
- También en el mismo código voy a calcular la precisón de esas predicciones.

```python
#1.- CARGO LA TABLA CSV y la almaceno en una variable llamada HOUSTON
import pandas as pd

HOUSTON = pd.read_csv('231206 housing data.csv', header=None)

# Pido una visualización rápida de los datos para ver su aspecto con el método to_string()
print('VISUALIZACIÓN RÁPIDA DE LOS DATOS')
print(HOUSTON.to_string())
print()

# Pido que me de la forma de esta tabla original
print('La forma de la matriz original de datos es:', HOUSTON.shape)
```

```python
# 2.- CONVIERTO el DataFrame HOUSTON en una matriz NumPy llamada data, y compruebo el tipo de elmentos.
data = HOUSTON.values 
print('La tabla HOUSTON es del tipo:',type(HOUSTON))
print('La tabla data es del tipo:', type(data))

# 3.- IDENTIFICO EN LA MATRIZ "data" EN ELEMENTOS DE ENTRADA Y SALIDA
# Los datos de salida (y) son la última columna
X, y = data[:, :-1], data[:, -1]

# 4.- COMPRUEBO LA FORMA DE LOS DOS ELEMENTOS DE ENTRADA Y SALIDA OBTENIDOS
print('La forma de los elementos de entrada y salida es', X.shape, y.shape)

# 5 .- DIVIDO LOS DATOS PARA EL TRAINTTEST (split into train test sets) 
# Para ello, primero importo: train_test_split
from sklearn.model_selection import train_test_split

# Luego tomo la decisión: el 67 por ciento se utilizará para entrenar el modelo y el 33 por ciento para evaluarlo.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=1)

# 6.- COMPRUEBO LA FORMA DE AMBAS DIVISIONES 
print('La forma para ambas divisones es:', X_train.shape, X_test.shape, y_train.shape, y_test.shape)

# 7.- AJUSTO/ENTRENO EL MODELO (fit the model) CON UN MODELO DE REGRESIÓN ALEATORIA EN EL CONJUNTO DE ENTRENAMIENTO:
# Primero importo la clase RandomForestRegressor desde el módulo sklearn.ensemble. 
# RandomForestRegressor es un modelo de regresión basado en la técnica de Random Forest (árboles de decisión). 
# se utiliza específicamente para resolver problemas de regresión, donde la variable de destino es continua.
# (si la variable hubiese sido discreta o categórica hubiese ultilizado "RandomForestClassifier" en su lugar)
from sklearn.ensemble import RandomForestRegressor

# Controlo la aleatorización durante la división con random_state
# El parámetro random_state se establece en 1, ==> se fijará la semilla para la generación de números aleatorios. 
# Fijar la semilla garantiza que los resultados sean reproducibles; es decir: 
# que si ejecuto el mismo código varias veces con la misma semilla, debería obtener los mismos resultados 
model = RandomForestRegressor(random_state=1)
model.fit(X_train, y_train)
# RandomForestRegressor crea el objeto que representa el modelo, mientras .fit()entrena o ajusta el modelo y lo devuelve

# 8.- HAGO PREDICCIONES:
# Primero tengo que importar mean_absolute_error
from sklearn.metrics import mean_absolute_error

yhat = model.predict(X_test)
# yhat = model.predict(X_test): Este código utiliza el modelo (model) previamente entrenado 
# para hacer predicciones sobre un conjunto de datos de prueba (X_test). 
# El resultado de estas predicciones se almacena en la variable yhat.

# 9.- EVALUO PREDICCIONES: 
# Para hacer predicciones y evaluarlas utilizo el mean absolute error (MAE / error medio absoluto).

mae = mean_absolute_error(y_test, yhat)
# "mae = mean_absolute_error(y_test, yhat)" calcula el error absoluto medio (MAE) entre las predicciones (yhat) 
# y las etiquetas reales (y_test). Explicado de forma más detallada:
# mean_absolute_error: calcula el error absoluto medio entre dos conjuntos de valores. La fórmula para el MAE es 
#la suma de las diferencias absolutas entre las predicciones y los valores reales, dividida por el número total de observaciones.
# El resultado (mae) me da una medida del error promedio absoluto de las predicciones en relación con las verdaderas etiquetas. 
# Un MAE más bajo indica que las predicciones están más cerca de los valores reales, 
# mientras que un MAE más alto indica mayores discrepancias entre las predicciones y las observaciones reales.

print('El resultado MAE (error absoluto medio) es : %.3f' % mae)

# En resumen, estas líneas de código realizan predicciones utilizando un modelo entrenado (model) 
# en un conjunto de datos de prueba (X_test), calculan la precisión de las predicciones utilizando 
# la función "mean_absolute_error" y finalmente imprimen la precisión en la consola. 
# La precisión es una medida importante para evaluar el rendimiento de un modelo de clasificación.
```

### 2.1.- ESTANDARIZACIÓN:
- Este método va bien cuando estamos utilizando modelos sensibles a la escala de las características (como la regresión lineal, máquinas de soporte vectorial, k-means, etc.). 
- según ChantGPT Puede resultar beneficioso estandarizar los datos numéricos para que tengan una media de 0 y una desviación estándar de 1.

```python
# Aplico el código de estandarización que aparece en el ScikitLearn Cheatsheet.pdf
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X_train)
standardized_X_train = scaler.transform(X_train)standardized_X_test = scaler.transform(X_test)
```

```python
import numpy as np
print("Conjunto de ENTRENAMIENTO estandarizado:\n")
print(np.round(standardized_X_train, 3))

print("\nConjunto de PRUEBA estandarizado:\n")
print(np.round(standardized_X_test, 3))
```

```python
# Comparo el GRUPO DE ENTRENAMIENTO antes y después de la estandarización.
print("Primeras 5 filas del GRUPO DE ENTRENAMIENTO originales:\n")
print(np.round(X_train[:5, :],3))

print("\nPrimeras 5 filas del GRUPO DE ENTRENAMIENTO estandarizados:\n")
print(np.round(standardized_X_train[:5, :],3))
```

```python
# SOBRE LAS TABLAS ESTANDARIZADAS VUELVO A COMPROBAR SUS FORMAS, A HACER PREDICCIONES 
# Y A COMPROBAR LA EFICACIA DE LAS PREDICCIONES: 

# 6.- COMPRUEBO LA FORMA DE AMBAS DIVISIONES 
print('La forma para ambas divisones es:', standardized_X_train.shape, standardized_X_test.shape, y_train.shape, y_test.shape)

# 7.- AJUSTO/ENTRENO EL MODELO (fit the model) CON UN MODELO DE REGRESIÓN ALEATORIA EN EL CONJUNTO DE ENTRENAMIENTO:
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(random_state=1)
model.fit(standardized_X_train, y_train)

# 8.- HAGO PREDICCIONES:
from sklearn.metrics import mean_absolute_error
yhat = model.predict(standardized_X_test)

# 9.- EVALUO PREDICCIONES: 
mae = mean_absolute_error(y_test, yhat)

print('El resultado MAE (error absoluto medio) es : %.3f' % mae)

# Observo que el resultado es el mismo que antes de estandarizar 
# ==> No alcanzo a entender el si la estandarización me ha beneficiado en algo.
```

__FINALMENTE__: hago la consulta a mi amigo CHATGPT, que me recomienda el siguiente código

```python
# CREAR COLUMNAS DUMMIES
X_train_df_dummy = pd.concat([X_train_df, pd.get_dummies(X_train_df['CHAS'], prefix='CHAS', drop_first=True)], axis=1)
X_test_df_dummy = pd.concat([X_test_df, pd.get_dummies(X_test_df['CHAS'], prefix='CHAS', drop_first=True)], axis=1)

# pd.get_dummies: función de pandas que crea columnas dummies a partir de la columna categórica'CHAS'. 
# (Crea una columna dummy para cada categoría única en 'CHAS').
# prefix='CHAS': Agrega el prefijo 'CHAS' a todas las nuevas columnas dummies para identificar su origen.
# drop_first=True: Se utiliza para evitar la multicolinealidad al eliminar la primera columna dummy ==> 
# esto significa que si hay dos categorías en 'CHAS', solo necesito una columna dummy para representarlas, 
# ya que la otra se puede inferir.
# pd.concat: Combina estas nuevas columnas dummies con el DataFrame original X_train_df y X_test_df, respectivamente.

# Elimino la columna original CHAS de las dos tablas, ya que ahora está representada por las dummies creadas.
X_train_df_dummy.drop('CHAS', axis=1, inplace=True)
X_test_df_dummy.drop('CHAS', axis=1, inplace=True)
```
